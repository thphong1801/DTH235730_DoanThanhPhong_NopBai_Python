```python
#Cau1
def CheckDoiXung(s):
    flag = True
    for i in range(len(s)):
        if s[i] != s[len(s) - i - 1]:
            flag = False
            break
    return flag

def main():
    print("Nhập 1 chuỗi:")
    s = input()
    if CheckDoiXung(s):
        print("Chuỗi bạn nhập đối xứng")
    else:
        print("Chuỗi bạn nhập không đối xứng")

while True:
    main()
    print("Tiếp không Thím? (c/k):")
    s = input()
    if s == "k":
        break
print("CÁM ƠN THÍM")
```

```python
#Cau2
def ToiUuChuoi(s):
    s2 = s.strip()          # bỏ khoảng trắng đầu và cuối
    arr = s2.split(' ')     # tách chuỗi theo khoảng trắng
    s2 = ""
    for x in arr:
        word = x
        if len(word.strip()) != 0:   # bỏ phần tử rỗng
            s2 = s2 + word + " "
    return s2.strip()       # bỏ khoảng trắng thừa cuối cùng

s = "   Trần   Duy   Thanh   "
print(s, "=>", len(s))
s = ToiUuChuoi(s)
print(s, "=>", len(s))
```

```python
#Cau3
def CheckPrime(x):
    if x < 2:
        return False
    dem = 0
    for i in range(1, x + 1):
        if x % i == 0:
            dem += 1
    return dem == 2

s = "5;7;8;-2;8;11;-13;9;10"
arr = s.split(';')

sochan = 0
soam = 0
sont = 0
tong = 0

for x in arr:
    print(x)
    number = int(x)
    if number % 2 == 0:
        sochan += 1
    if number < 0:
        soam += 1
    if CheckPrime(number):
        sont += 1
    tong += number

print("Số chẵn =", sochan)
print("Số âm =", soam)
print("Số nguyên tố =", sont)
print("Trung bình =", tong / len(arr))
```

```python
#Cau5
def DemChuoi(s):
    inhoa = 0
    inthuong = 0
    chu_so = 0
    kytu_db = 0
    khoang_trang = 0
    nguyen_am = 0
    phu_am = 0

    # Tập hợp nguyên âm (bao gồm cả chữ hoa)
    nguyen_am_set = "aeiouAEIOUáàảãạăắằẳẵặâấầẩẫậóòỏõọôốồổỗộơớờởỡợíìỉĩịúùủũụưứừửữựéèẻẽẹêếềểễệ"

    for ch in s:
        if ch.isupper():
            inhoa += 1
        elif ch.islower():
            inthuong += 1
        elif ch.isdigit():
            chu_so += 1
        elif ch.isspace():
            khoang_trang += 1
        else:
            kytu_db += 1

        if ch.isalpha():  # nếu là chữ cái
            if ch in nguyen_am_set:
                nguyen_am += 1
            else:
                phu_am += 1

    print("Số chữ IN HOA:", inhoa)
    print("Số chữ thường:", inthuong)
    print("Số chữ là chữ số:", chu_so)
    print("Số ký tự đặc biệt:", kytu_db)
    print("Số ký tự khoảng trắng:", khoang_trang)
    print("Số chữ là Nguyên Âm:", nguyen_am)
    print("Số chữ là Phụ Âm:", phu_am)

# --- Chương trình chính ---
s = input("Nhập 1 chuỗi: ")
DemChuoi(s)
```

```python
#Cau6
def NegativeNumberInStrings(s):
    numbers = []
    i = 0
    while i < len(s):
        # Nếu ký tự hiện tại là '-' và sau nó là một chữ số
        if s[i] == '-' and i + 1 < len(s) and s[i + 1].isdigit():
            num = '-'
            i += 1
            # Lấy tất cả các chữ số sau dấu '-'
            while i < len(s) and s[i].isdigit():
                num += s[i]
                i += 1
            numbers.append(int(num))
        else:
            i += 1
    return numbers

# --- Chương trình chính ---
chuoi = input("Nhập chuỗi: ")
ketqua = NegativeNumberInStrings(chuoi)
if ketqua:
    print("Các số nguyên âm trong chuỗi là:", ketqua)
else:
    print("Không có số nguyên âm nào trong chuỗi.")
```

```python
#Cau7
def toi_uu_chuoi_danh_tu(s):
    # Loại bỏ khoảng trắng đầu và cuối, tách các từ bằng split()
    tu = s.strip().split()

    # Chuẩn hóa từng từ: viết hoa chữ đầu, còn lại viết thường
    tu_chuan = [t.capitalize() for t in tu]

    # Ghép lại thành một chuỗi, cách nhau bằng 1 khoảng trắng
    ket_qua = " ".join(tu_chuan)

    return ket_qua

# --- Chạy thử ---
s = "   TRần   duY    thAnH   "
print(toi_uu_chuoi_danh_tu(s))
```

```python
#Cau8
import os

# Hàm 1: Lấy tên file (kèm phần mở rộng)
def lay_ten_file(duong_dan):
    # Chuẩn hóa đường dẫn để tránh lỗi giữa "\" và "/"
    duong_dan = duong_dan.strip().replace("/", "\\")
    return os.path.basename(duong_dan)

# Hàm 2: Lấy tên bài hát (không kèm phần mở rộng)
def lay_ten_bai_hat(duong_dan):
    duong_dan = duong_dan.strip().replace("/", "\\")
    return os.path.splitext(os.path.basename(duong_dan))[0]

# --- Chạy thử ---
ds_duong_dan = [
    r"d:\music\muabui.mp3",
    r"E:/nhac/viet/thangtu.mp3",
    r"C:\Users\Admin\Downloads\mua ha nam ay.wav",
    r"   F:\Nhac Tre\yeuem.flac   "
]

for duong_dan in ds_duong_dan:
    print("Đường dẫn:", duong_dan)
    print("→ Tên file:", lay_ten_file(duong_dan))
    print("→ Tên bài hát:", lay_ten_bai_hat(duong_dan))
    print()

```
