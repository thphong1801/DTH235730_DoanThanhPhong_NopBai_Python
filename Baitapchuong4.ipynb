```python
#Cau4
def ROI(dt, cp):
    """H√†m t√≠nh t·ª∑ l·ªá ROI = (Doanh thu - Chi ph√≠) / Chi ph√≠"""
    return (dt - cp) / cp

def GoiYDauTu(roi):
    """H√†m ƒë∆∞a ra g·ª£i √Ω ƒë·∫ßu t∆∞ d·ª±a tr√™n ROI"""
    if roi >= 0.75:
        return "N√™n ƒë·∫ßu t∆∞"
    else:
        return "Kh√¥ng n√™n ƒë·∫ßu t∆∞"

print("=== CH∆Ø∆†NG TR√åNH T√çNH ROI ===")

try:
    dt = float(input("Nh·∫≠p Doanh Thu: "))
    cp = float(input("Nh·∫≠p Chi Ph√≠: "))

    if cp <= 0:
        print("‚ùå L·ªói: Chi ph√≠ ph·∫£i l·ªõn h∆°n 0!")
    elif dt < 0:
        print("‚ùå L·ªói: Doanh thu kh√¥ng ƒë∆∞·ª£c √¢m!")
    else:
        roi = ROI(dt, cp)
        print(f"\nüìä T·ªâ l·ªá ROI = {roi:.2f} ({roi*100:.1f}%)")
        print("üí° G·ª£i √Ω:", GoiYDauTu(roi))

except ValueError:
    print("‚ùå Vui l√≤ng nh·∫≠p s·ªë h·ª£p l·ªá!")
```

```python
#Cau5
def fibonacci(n):
    if n <= 2:
        return 1
    f1, f2 = 1, 1
    for i in range(3, n + 1):
        f1, f2 = f2, f1 + f2
    return f2

def listfibo(n):
    for i in range(1, n + 1):
        print(fibonacci(i), end='\t')

print(f"F(9) = {fibonacci(9)}")
print("D√£y Fibonacci 1 ‚Üí 9:")
listfibo(9)
```

```python
#Cau7
import math

def distance(A, B):
    """Tr·∫£ v·ªÅ kho·∫£ng c√°ch gi·ªØa 2 ƒëi·ªÉm A v√† B.
    A, B l√† tuple (x, y)."""
    xA, yA = A
    xB, yB = B
    return math.sqrt((xB - xA)**2 + (yB - yA)**2)

def read_point(name):
    """ƒê·ªçc to·∫° ƒë·ªô 1 ƒëi·ªÉm t·ª´ input, tr·∫£ v·ªÅ (x, y) d∆∞·ªõi d·∫°ng float."""
    while True:
        s = input(f"Nh·∫≠p to·∫° ƒë·ªô ƒëi·ªÉm {name} theo d·∫°ng x,y (v√≠ d·ª• 1.5,2): ").strip()
        try:
            parts = s.split(',')
            if len(parts) != 2:
                raise ValueError
            x = float(parts[0].strip())
            y = float(parts[1].strip())
            return (x, y)
        except ValueError:
            print("Sai ƒë·ªãnh d·∫°ng. Vui l√≤ng nh·∫≠p l·∫°i nh∆∞: 3,4 ho·∫∑c 1.2, -5")

# Ch∆∞∆°ng tr√¨nh ch√≠nh
print("=== T√çNH ƒê·ªò D√ÄI ƒêO·∫†N TH·∫≤NG AB ===")
A = read_point('A')
B = read_point('B')

d = distance(A, B)
print(f"\nTo·∫° ƒë·ªô A = ({A[0]}, {A[1]})")
print(f"To·∫° ƒë·ªô B = ({B[0]}, {B[1]})")
print(f"ƒê·ªô d√†i ƒëo·∫°n AB = {d:.4f}")
```

```python
#Cau11
def sum1(n):
    s = 0
    while n > 0:
        s += 1
        n -= 1
    return s

def sum2():
    global val
    s = 0
    while val > 0:
        s += 1
        val -= 1
    return s

def sum3():
    s = 0
    for i in range(val, 0, -1):
        s += 1
    return s
def main():
    print("truong hop 1")
    global val
    val = 5
    print(sum1(5))
    print(sum2())
    print(sum3())
main()
def main():
    print("truong hop 2")
    global val
    val = 5
    print(sum1(5))
    print(sum3())
    print(sum2())
main()
def main():
    print("truong hop 3")
    global val
    val = 5
    print(sum2())
    print(sum1(5))
    print(sum3())
main()
```

```python
#Cau12
def oscillate(start, count):
    n = start
    for i in range(count):
        yield n            # Tr·∫£ v·ªÅ gi√° tr·ªã hi·ªán t·∫°i
        yield -n           # Tr·∫£ v·ªÅ gi√° tr·ªã ƒë·ªëi x·ª©ng c·ªßa n√≥
        # C·∫≠p nh·∫≠t n: n·∫øu n √¢m th√¨ ti·∫øn v·ªÅ 0 (n + 1), n·∫øu n d∆∞∆°ng th√¨ l√πi v·ªÅ 0 (n - 1)
        n += 1 if n < 0 else -1


# --- Ch∆∞∆°ng tr√¨nh ch√≠nh ---
for n in oscillate(-3, 5):
    print(n, end=' ')
print()
```

```python
#cau13
# C√¢u 13: H√†m ki·ªÉm tra s·ªë ho√†n thi·ªán, s·ªë th·ªãnh v∆∞·ª£ng

# H√†m t√≠nh t·ªïng c√°c ∆∞·ªõc s·ªë c·ªßa n (kh√¥ng t√≠nh ch√≠nh n)
def tong_uoc(n):
    s = 0
    for i in range(1, n):
        if n % i == 0:
            s += i
    return s

# a) Ki·ªÉm tra s·ªë ho√†n thi·ªán
def la_so_hoan_thien(n):
    return tong_uoc(n) == n

# b) Ki·ªÉm tra s·ªë th·ªãnh v∆∞·ª£ng
def la_so_thinh_vuong(n):
    return tong_uoc(n) > n

# --- Ch∆∞∆°ng tr√¨nh ch√≠nh ---
n = int(input("Nh·∫≠p s·ªë nguy√™n d∆∞∆°ng n: "))

if la_so_hoan_thien(n):
    print(f"{n} l√† s·ªë ho√†n thi·ªán.")
elif la_so_thinh_vuong(n):
    print(f"{n} l√† s·ªë th·ªãnh v∆∞·ª£ng.")
else:
    print(f"{n} kh√¥ng ph·∫£i l√† s·ªë ho√†n thi·ªán hay th·ªãnh v∆∞·ª£ng.")
```
