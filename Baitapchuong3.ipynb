```python
#cau6
def docso(n):

  donvi=["","mot","hai","ba","nam","sau","bay","tam","chin"]
  if n<10:
    return donvi[n]
  elif n<20:
    if n==10:
      return "Muoi"
    elif n==15:
      return "muoi lam"
    else:
      return "Muoi" + donvi[n%10]
  else:
    chuc = n//10
    dv =n % 10
    ketqua = donvi[chuc]+ " muoi "
    if dv>0:
      if dv==1:
        ketqua += " mot"
      elif dv==5:
        ketqua += "lam"
      else:
        ketqua +=""+ donvi[dv]
    return ketqua
n= int(input("Nhap so n (be hon 99) : "))
print("doc so:", docso(n))
```

```python
#cau7
def ngaytrongthang(thang,nam):

  if thang == 2:
    if(nam % 400 ==0) or (nam % 4 == 0 and nam % 100!= 0):
      return 29
    else:
      return 28
  elif  thang in [1,3,5,7,8,10,12]:
    return 31
  else:
    return 30
def nextday(ngay,thang,nam):
  ngay +=1
  if ngay >ngaytrongthang(thang,nam):
    ngay=1
    thang+=1
    if thang>12:
      thang=1
      nam+=1
  return ngay,thang,nam
d,m,y= map(int,input("Nhap ngay thang nam: ").split("/"))
nd,nm,ny=nextday(d,m,y)
print(f"Ngay Ke tiep: {nd:02d}/{nm:02d}/{ny:02d}")
```

```python
#cau18
n = int(input("Nhập chiều cao n: "))

# Hình 1: Hình chữ nhật rỗng
print("Hình 1:")
for i in range(n):
    for j in range(n):
        if i == 0 or i == n-1 or j == 0 or j == n-1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()

# Hình 2: Tam giác vuông góc dưới trái
print("Hình 2:")
for i in range(1, n+1):
    for j in range(i):
        print("*", end=" ")
    print()

# Hình 3: Tam giác vuông góc trên trái
print("Hình 3:")
for i in range(n, 0, -1):
    for j in range(i):
        print("*", end=" ")
    print()

# Hình 4: Tam giác cân (giữa)
print("Hình 4:")
for i in range(n):
    for j in range(n - i - 1):
        print(" ", end=" ")
    for j in range(2 * i + 1):
        print("*", end=" ")
    print()
```

```python
#Cau19
import math

def S_truyhoi(x, n):
    total = 0.0
    T = x  # T_0
    for k in range(n+1):
        total += T
        # chuẩn bị T_{k+1} từ T_k
        denom = (2*k + 2) * (2*k + 3)   # (2k+2)(2k+3)
        T = T * (x * x) / denom
    return total

# Ví dụ
x = 1.0
n = 3
print("S =", S_truyhoi(x, n))   # ~ 1.1751984126984128
```
