```python
#Cau1
from random import randrange

print("Chương trình xử lý List")
n = int(input("Nhập số phần tử: "))

# Tạo list ngẫu nhiên
lst = [randrange(-100, 100) for _ in range(n)]

print("List ngẫu nhiên là:")
print(lst)

# Thêm phần tử mới
value = int(input("Mời bạn thêm số mới: "))
lst.append(value)
print("List sau khi thêm:")
print(lst)

# Đếm số xuất hiện
k = int(input("Bạn muốn đếm số nào: "))
dem = lst.count(k)
print(k, "xuất hiện", dem, "lần trong list")

# Hàm kiểm tra số nguyên tố
def CheckPrime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Đếm và tính tổng các số nguyên tố
demnt = 0
tongnt = 0
for x in lst:
    if CheckPrime(x):
        demnt += 1
        tongnt += x

print("Có", demnt, "số nguyên tố trong list")
print("Tổng =", tongnt)

# Sắp xếp list
lst.sort()
print("List sau khi sort:")
print(lst)

# Xóa list
del lst
print("List sau khi xóa:")
try:
    print(lst)
except NameError:
    print("List đã bị xóa khỏi bộ nhớ!")

```

```python
#Cau2
from random import randrange

lst = []   # tạo list rỗng

print("Nhập số phần tử:")
n = int(input())

# Tạo list ngẫu nhiên
for i in range(n):
    lst.append(randrange(0, 100))

print("List sau khi tạo ngẫu nhiên là:")
print(lst)

# Chèn thêm phần tử mới
x = int(input("Mời bạn chèn thêm số mới: "))
lst.append(x)
print("List sau khi chèn:")
print(lst)

# Xóa tất cả các phần tử có giá trị k
k = int(input("Mời nhập số để xóa: "))
while lst.count(k) > 0:
    lst.remove(k)

print("List sau khi xóa:")
print(lst)

# Hàm kiểm tra list có đối xứng hay không
def CheckDoiXung(lst):
    for i in range(len(lst) // 2):   # chỉ cần duyệt nửa đầu
        if lst[i] != lst[len(lst) - i - 1]:
            return False
    return True

# Kiểm tra và in kết quả
if CheckDoiXung(lst):
    print("List đối xứng")
else:
    print("List không đối xứng")
```

```python
#Cau3
from random import randrange

# Tạo ma trận m x n với các số ngẫu nhiên từ 0 đến 99
def TaoMaTran(m, n):
    D = []
    for i in range(m):
        row = []
        for j in range(n):
            row.append(randrange(100))
        D.append(row)
    return D

# Xuất ma trận
def XuatMaTran(D):
    for row in D:
        for element in row:
            print(element, end='\t')
        print()

# Lấy dòng r của ma trận
def LayDong(D, r):
    return D[r]

# Xuất list 1 chiều
def XuatList1Chieu(R):
    for element in R:
        print(element, end='\t')
    print()  # xuống dòng sau khi in xong

# Lấy cột c của ma trận
def LayCot(D, c):
    C = []
    for i in range(len(D)):
        C.append(D[i][c])
    return C

# Tìm số lớn nhất trong ma trận
def MAX(D):
    max_val = D[0][0]
    for i in range(len(D)):
        for j in range(len(D[i])):
            if max_val < D[i][j]:
                max_val = D[i][j]
    return max_val

# ================== Chương trình chính ==================
print("Nhập số dòng:")
m = int(input())
print("Nhập số cột:")
n = int(input())

D = TaoMaTran(m, n)
print("Ma trận vừa tạo:")
XuatMaTran(D)

print("Mời bạn nhập dòng muốn xuất:")
r = int(input())
print(f"Dòng {r} là:")
XuatList1Chieu(LayDong(D, r))

print("Mời bạn nhập cột muốn xuất:")
c = int(input())
print(f"Cột {c} là:")
XuatList1Chieu(LayCot(D, c))

max_val = MAX(D)
print("Số lớn nhất trong ma trận =", max_val)
```

```python
#Cau6
import random

N = int(input("Nhập số lượng phần tử N: "))
lst = []

while len(lst) < N:
    x = random.randint(0, 100)
    if x not in lst:
        lst.append(x)

print("Danh sách ngẫu nhiên không trùng nhau là:")
print(lst)
```

```python
#Cau7
N = int(input("Nhập số lượng phần tử: "))
lst = []

print("Nhập các số theo thứ tự tăng dần:")

for i in range(N):
    while True:
        x = int(input(f"Nhập phần tử thứ {i+1}: "))
        # Nếu là phần tử đầu tiên hoặc lớn hơn phần tử trước đó
        if i == 0 or x >= lst[-1]:
            lst.append(x)
            break
        else:
            print("❌ Sai quy tắc! Số phải >= số trước đó. Vui lòng nhập lại.")

print("\n✅ Dãy số sau khi nhập xong:")
print(lst)
```

```python
#Cau8
# Nhập số lượng phần tử
n = int(input("Nhập số lượng phần tử n: "))

# Nhập các số thực vào danh sách
M = []
for i in range(n):
    x = float(input(f"Nhập M[{i}]: "))
    M.append(x)

# Sắp xếp giảm dần
M.sort(reverse=True)

# In ra kết quả
print("\n✅ Dãy số sau khi sắp xếp giảm dần:")
print(M)
```

```python
#Cau9
# Hàm kiểm tra số nguyên tố
def la_nguyen_to(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Nhập mảng
M = list(map(int, input("Nhập các số tự nhiên (cách nhau bằng dấu cách): ").split()))

# Lọc ra các loại số
le = [x for x in M if x % 2 != 0]
chan = [x for x in M if x % 2 == 0]
nguyen_to = [x for x in M if la_nguyen_to(x)]
khong_nt = [x for x in M if not la_nguyen_to(x)]

# In kết quả
print("\nDòng 1: Các số lẻ:", le, f"→ Có {len(le)} số lẻ")
print("Dòng 2: Các số chẵn:", chan, f"→ Có {len(chan)} số chẵn")
print("Dòng 3: Các số nguyên tố:", nguyen_to)
print("Dòng 4: Các số không phải số nguyên tố:", khong_nt)
```

```python
#Cau10
# Hàm nhập ma trận
def nhap_ma_tran(ten):
    rows = int(input(f"Nhập số hàng của ma trận {ten}: "))
    cols = int(input(f"Nhập số cột của ma trận {ten}: "))
    print(f"Nhập các phần tử của ma trận {ten}:")
    M = []
    for i in range(rows):
        row = list(map(float, input(f"Hàng {i+1}: ").split()))
        M.append(row)
    return M

# Hàm in ma trận
def in_ma_tran(M, name="Ma trận"):
    print(f"\n{name}:")
    for row in M:
        print(row)

# Hàm cộng hai ma trận
def cong_ma_tran(A, B):
    return [[A[i][j] + B[i][j] for j in range(len(A[0]))] for i in range(len(A))]

# Hàm hoán vị ma trận
def hoan_vi(M):
    return [[M[j][i] for j in range(len(M))] for i in range(len(M[0]))]

# --- Chương trình chính ---
A = nhap_ma_tran("A")
B = nhap_ma_tran("B")

# Cộng hai ma trận
C = cong_ma_tran(A, B)

# Hoán vị
AT = hoan_vi(A)
BT = hoan_vi(B)

# In kết quả
in_ma_tran(A, "Ma trận A")
in_ma_tran(B, "Ma trận B")
in_ma_tran(C, "Tổng A + B")
in_ma_tran(AT, "Hoán vị của A")
in_ma_tran(BT, "Hoán vị của B")
```
